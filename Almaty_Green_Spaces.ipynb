```python
import osmnx as ox
import matplotlib.pyplot as plt

# Получение полигона Алматы
admin = ox.geocode_to_gdf('Almaty')
admin_poly = admin.geometry.values[0]

# Получение POI зеленых насаждений с расширенным набором тегов
tags = {
    'leisure': ['park', 'garden', 'playground', 'golf_course', 'nature_reserve'],
    'landuse': ['forest', 'grass', 'meadow', 'recreation_ground', 'village_green', 'orchard', 'vineyard'],
    'natural': ['wood', 'tree_row', 'scrub', 'heath', 'grassland']
}

green_areas = ox.geometries_from_polygon(admin_poly, tags=tags)

# Вычисление площади зеленых насаждений
green_areas = green_areas.to_crs(epsg=3857)  # Преобразование в метры для корректного вычисления площади
green_areas['area'] = green_areas.geometry.area
total_green_area = green_areas['area'].sum()

# Вычисление общей площади города
admin = admin.to_crs(epsg=3857)
city_area = admin.geometry.area.values[0]

# Вычисление процента зеленых насаждений
green_percentage = (total_green_area / city_area) * 100

# Визуализация
plt.style.use('dark_background')
f, ax = plt.subplots(1, figsize=(10, 10))
green_areas.plot(ax=ax, color='green', linewidth=0.3, markersize=0.3)
ax.set_title(f"Green Spaces in Almaty\nGreen Coverage: {green_percentage:.2f}%", fontsize=20, color='white')
ax.axis('off')

# Сохранение визуализации
output_path = 'Almaty_green_spaces_percent.png'
f.savefig(output_path, dpi=300, bbox_inches='tight', facecolor=f.get_facecolor())

# Загрузка файла
from google.colab import files
files.download(output_path)

# Вывод процента зеленых насаждений
print(f"Percentage of green spaces in Almaty: {green_percentage:.2f}%")
```
