```python
import osmnx as ox
import matplotlib.pyplot as plt

# Получение полигона Алматы
admin = ox.geocode_to_gdf('Almaty')
admin_poly = admin.geometry.values[0]

# Получение POI зеленых насаждений с расширенным набором тегов
tags = {
    'leisure': ['park', 'garden', 'playground', 'golf_course', 'nature_reserve'],
    'landuse': ['forest', 'grass', 'meadow', 'recreation_ground', 'village_green', 'orchard', 'vineyard'],
    'natural': ['wood', 'tree_row', 'scrub', 'heath', 'grassland']
}

green_areas = ox.geometries_from_polygon(admin_poly, tags=tags)

# Визуализация
plt.style.use('dark_background')
f, ax = plt.subplots(1, figsize=(10, 10))
green_areas.plot(ax=ax, color='green', linewidth=0.3, markersize=0.3)
ax.set_title("Green Spaces in Almaty", fontsize=20, color='white')
ax.axis('off')

# Сохранение визуализации
output_path = 'Almaty_green_spaces.png'
f.savefig(output_path, dpi=300, bbox_inches='tight', facecolor=f.get_facecolor())

# Загрузка файла
from google.colab import files
files.download(output_path)
```

```python
import osmnx as ox
import matplotlib.pyplot as plt

# Получение полигона Алматы
admin = ox.geocode_to_gdf('Almaty')
admin_poly = admin.geometry.values[0]

# Определение всех возможных тегов, связанных с зелеными насаждениями
tags = {
    'leisure': ['park', 'garden', 'playground', 'golf_course', 'nature_reserve', 'common', 'pitch', 'dog_park'],
    'landuse': ['forest', 'grass', 'meadow', 'recreation_ground', 'village_green', 'orchard', 'vineyard', 'allotments', 'cemetery'],
    'natural': ['wood', 'tree_row', 'scrub', 'heath', 'grassland', 'fell', 'marsh', 'wetland', 'bare_rock', 'scree', 'shingle', 'sand', 'mud'],
    'boundary': ['protected_area']
}

# Получение POI зеленых насаждений
green_areas = ox.geometries_from_polygon(admin_poly, tags=tags)

# Визуализация
plt.style.use('dark_background')
f, ax = plt.subplots(1, figsize=(10, 10))
green_areas.plot(ax=ax, color='green', linewidth=0.3, markersize=0.3)
ax.set_title("Green Spaces in Almaty", fontsize=20, color='white', fontname='Arial')
ax.axis('off')

# Сохранение визуализации
output_path = 'Almaty_green_spaces.png'
f.savefig(output_path, dpi=300, bbox_inches='tight', facecolor=f.get_facecolor())

# Загрузка файла
from google.colab import files
files.download(output_path)
```
